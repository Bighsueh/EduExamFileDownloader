```python
from ExamDataScraper import ExamDataScraper
import pandas as pd
```

```python
all_dfs = []

for page in range(1, 2001):
	if page % 100 == 0:
		print(f'Current iteration: {page}')
	scraper = ExamDataScraper('https://exam.naer.edu.tw', '/searchResult.php', page=page)
	df = scraper.get_dataframe()
	all_dfs.append(df)

combined_df = pd.concat(all_dfs, ignore_index=True)
```

```python
combined_df.to_csv('exam_data.csv', index=False, encoding='utf-8-sig')
```
